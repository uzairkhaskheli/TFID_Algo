In this assignment, you are provided with a file `Article.txt'. This file contains an article about covid-19 from CBC newspaper. The goal of this assignment is to summarize automatically the given text using **TF-IDF** algorithms.

```python
#importing math library:
import math
text = ""
with open('Article.txt', encoding="utf-8") as file:
    for line in file:
        text += line
```

```python
#split the text with . 
lines = text.lower().split(".")
```

```python

words_in_lines = {}
lines_counter = []
counter = 0
#read lines
for line in lines:
    words_lines = {}
    lines_counter.append(line)
    words_in_line = []
    
    line_split = line.split(" ")
    
    for words in line_split:
        words_in_line.append(words)
    
    words_lines[line.strip()] = words_in_line
    words_in_lines[f"line-{counter}"] = words_lines
    counter = counter + 1
```

```python
# For each word compute how many sentences contain that word.
number_of_sentences_in_given_text = len(lines_counter)
```

```python
number_of_sentences_in_given_text
```

```python
words_in_lines
```

```python
words_frequency = {}
```

```python
#introducing a function to get unique words 
def getUniqueWords(allWords):
    uniqueWords = [] 
    for i in allWords:
        if not i in uniqueWords:
            uniqueWords.append(i)
    return uniqueWords
```

```python
for dictionary_keys, dictionary_values in words_in_lines.items():    
    words_frequency_dictionary = {}
    for keys, words_in_keys in dictionary_values.items():
        unqiue_words = getUniqueWords(words_in_keys)
        freqs = {}
        for words in unqiue_words:
            freqs[words] = words_in_keys.count(words)            
            words_frequency_dictionary[keys] = freqs
        
    words_frequency[dictionary_keys] = words_frequency_dictionary
```

```python
words_frequency
```

Create a matrix ***termFrequency***: The termFrequency matrix should
be a dictionary where each key is a sentence and the value is also a
dictionary of word frequency.

```python
#created the term frequency matrix
matrix_term_frequency = {}
TF = []

#creating dictionary where the key is a sentence and the value is a again a dictionary of work freuqency
for dictionary_keys, dictionary_values in words_frequency.items(): 
    sentence_keys = {}
    
    for nkeys, nvalues in dictionary_values.items():
        total_number_of_words_in_sentence = 0
        
        term_frequency_keys = {}
        
        for item in nvalues.items():
            total_number_of_words_in_sentence = total_number_of_words_in_sentence + item[1]
             
        for item in nvalues.items():
            word_identifier = {} 
            word_counter = {}
            term_frequency = {}
            
            term_freq = item[1] / total_number_of_words_in_sentence 
            TF.append(term_freq)
            term_frequency[f"term-frequency"] = term_freq       
            word_counter[f"total-count-{total_number_of_words_in_sentence}"] = term_frequency
            word_identifier[f"count-{item[1]}"] = word_counter   
            term_frequency_keys[f"{item[0]}"] = word_identifier     
            
        sentence_keys[f"{nkeys}"] = term_frequency_keys

    matrix_term_frequency[dictionary_keys] = sentence_keys
```

```python
matrix_term_frequency
```

```python
words_in_sentences = {}
for dictionary_keys, dictionary_values in words_frequency.items(): 
    for nkeys, nvalues in dictionary_values.items():        
        for item in nvalues.items():
            word = item[0]
            words_counter = {}
            for sentence in lines_counter:
                counter = 0               
                if word in sentence:
                    counter = counter + 1
                    words_counter[sentence] = f"{word}-count-{counter}"
            words_in_sentences[word] = words_counter
```

For each word compute how many sentences contain that ***word***.

```python
number_of_sentences_with_term_word_in_it = []

for keys, values in words_in_sentences.items():
    number_of_sentences_with_term_word_in_it.append(len(values))
```

```python
number_of_sentences_with_term_word_in_it
```

Calculate IDF for each word in a sentence. 
IDF(word) = log e(Total number of sentences/ number of
sentences with term word in it)

```python
IDF = []
for value in number_of_sentences_with_term_word_in_it:
    IDF.append(math.log(number_of_sentences_in_given_text / value))
```

```python
IDF
```

```python
TF
```

```python
#Compute the TF-IDF for each word in each sentence.
TF_IDF = [element_in_TF * element_in_IDF for element_in_TF, element_in_IDF in zip(TF,IDF)]
TF_IDF
```

Use the TF-IDF computed in (7) and give a weight for each sentence.

```python
weight = [element_in_TF * math.log(number_of_sentences_in_given_text / element_in_IDF) for element_in_TF, element_in_IDF in zip(TF,IDF)]
weight
```

```python
#Threshold: compute the average sentence weight
average_weight = sum(weight) / len(weight)
average_weight
```
